One application of k-means is image compression. Image compression allows to reduce the number of colors in an image to a specified number of clusters (K). In this example, the machine learning OpenCV is used. The compressed image consists of the main 4 colors because parameter K is set to 4. 

```python
#import necessary modules
import cv2 as cv
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#load image and convert it to matrix with n_rows * n_cols x 3 dimensions
img = skimage.io.imread('mandarin.jpg')
Z = img.reshape((-1, 3))

# convert to np.float32
Z = np.float32(Z)

# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv.TERM_CRITERIA_MAX_ITER, 10, 1)
K = 4
compactness,labels,centers =cv.kmeans(Z,K,None,criteria,1,cv.KMEANS_RANDOM_CENTERS)

# Now convert into uint8, and make original image
center = np.uint8(centers)
flat_img = center[labels.flatten()]
compressed_img = flat_img.reshape((img.shape))

# Determine the figures size
fig = plt.figure(figsize=(15,15))
plt.subplot(1, 2, 1) #first (original) image
plt.imshow(img)

plt.subplot(1, 2, 2) #second (compressed) image
plt.imshow(compressed_img)
plt.show()
```
